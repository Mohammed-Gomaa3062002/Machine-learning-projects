## Feature Scaling

```
data.columns
```

```
data_copy = data.copy()
```

```
# Rescale numerical features using Min-Max Scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_copy[['Density', 'BodyFat', 'Age', 'Weight', 'Height', 'Neck', 'Chest',
       'Abdomen', 'Hip', 'Thigh', 'Knee', 'Ankle', 'Biceps', 'Forearm',
       'Wrist']] = scaler.fit_transform(data_copy[['Density', 'BodyFat', 'Age', 'Weight', 'Height', 'Neck', 'Chest',
       'Abdomen', 'Hip', 'Thigh', 'Knee', 'Ankle', 'Biceps', 'Forearm',
       'Wrist']])
```

## Splitting the dataset into the Training set and Test set

```
# Split the dataset into training and testing sets
X = data_copy.iloc[:,:-1]
y = data_copy.iloc[:,-1]
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

# Training the Linear Regression on the Training set

```
from sklearn.linear_model import LinearRegression
# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

# Evalution Model

```
from sklearn.metrics import mean_squared_error, r2_score
lin_reg_pred = lin_reg.predict(X_test)
lin_reg_mse = mean_squared_error(y_test, lin_reg_pred)
print("Linear Regression MSE:", lin_reg_mse)
print("R2 Score:",r2_score(y_test, lin_reg_pred))
```

# Training the DecisionTreeRegressor model on the Training set

```
# Decision Tree Regression
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor()
dt_reg.fit(X_train, y_train)
```

```
dt_reg_pred = dt_reg.predict(X_test)
dt_reg_mse = mean_squared_error(y_test, dt_reg_pred)
print("Decision Tree Regression MSE:", dt_reg_mse)
print("R2 Score:",r2_score(y_test, dt_reg_pred))
```

# Training the Random Forest Regressor model on the Training set

```
# Random Forest Regression
from sklearn.ensemble import RandomForestRegressor

rf_reg = RandomForestRegressor()
rf_reg.fit(X_train, y_train)
rf_reg_pred = rf_reg.predict(X_test)
rf_reg_mse = mean_squared_error(y_test, rf_reg_pred)
print("Random Forest Regression MSE:", rf_reg_mse)
print("R2 Score:",r2_score(y_test, rf_reg_pred))
```

#Visualize Scories for Models

```
# Draw acc
# Create a dataframe to hold accuracies
accuracies_df = pd.DataFrame({'Model': ['lin_reg', 'dt_reg','rf_reg'],
                              'MSE': [lin_reg_mse, dt_reg_mse,rf_reg_mse]})

# Draw a bar plot
plt.figure(figsize=(5, 5))
sns.barplot(x='Model', y='MSE', data=accuracies_df, palette='Set2')
plt.title('Mean Abs')
plt.ylim(0, 0.03)  # Set the y-axis limit to better visualize differences
plt.ylabel('mean Absolute Error')
plt.xlabel('Model')
plt.show()
```
